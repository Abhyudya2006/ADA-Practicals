```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

def plot_greedy_knapsack_results():
    try:
        df = pd.read_csv("greedy_knapsack_runtime.csv")
    except FileNotFoundError:
        print("Error: 'greedy_knapsack_runtime.csv' not found.")
        return

    # Extract columns safely
    sizes = df.iloc[:, 0].values
    times = df.iloc[:, 1].values

    # Compute n log n upper bound
    OFFSET = 1.15
    c = (times[-1] / (sizes[-1] * np.log2(sizes[-1]))) * OFFSET
    bound_nlogn = c * (sizes * np.log2(sizes))

    plt.style.use("bmh")
    plt.figure(figsize=(10, 6))

    plt.plot(sizes, times,
             marker='o',
             linewidth=2,
             label="Measured Time")

    plt.plot(sizes, bound_nlogn,
             linestyle="--",
             linewidth=2,
             label="Upper Bound (n log n)")

    plt.title("Greedy Fractional Knapsack Performance")
    plt.xlabel("Number of Items (n)")
    plt.ylabel("Execution Time (seconds)")
    plt.legend()
    plt.grid(True, linestyle=":", alpha=0.6)
    plt.tight_layout()

    plt.show()

# 🔥 IMPORTANT — call the function
plot_greedy_knapsack_results()
```
