```python
import csv
import matplotlib.pyplot as plt
import math

plt.style.use('seaborn-v0_8')   # clean modern style

x = []
y = []
nlogn = []

# Read CSV
with open("activity_selection_analysis_test.csv", "r") as file:
    reader = csv.reader(file)
    next(reader)

    for row in reader:
        n = int(row[0])
        t = float(row[1])

        x.append(n)
        y.append(t)
        nlogn.append(n * math.log2(n))

# Normalize n log n curve
scale = (y[-1] / nlogn[-1]) * 1.3
nlogn = [val * scale for val in nlogn]

# Plot
plt.figure(figsize=(10,6))

plt.plot(x, y, marker='o', linewidth=2.5, color='#1f77b4', label="Measured Time")

plt.plot(x, nlogn, linestyle='--', linewidth=2.5, color='#d62728', label="Upper Bound (n log n)")

plt.xlabel("Number of Activities (n)", fontsize=12)
plt.ylabel("Execution Time (seconds)", fontsize=12)
plt.title("Activity Selection Performance Analysis", fontsize=14, weight='bold')

plt.legend()
plt.grid(True, linestyle='--', alpha=0.5)

plt.tight_layout()
plt.show()
```

```python
import csv
import matplotlib.pyplot as plt

x = []
selected = []

with open("activity_selection_count.csv", "r") as file:
    reader = csv.reader(file)
    next(reader)

    for row in reader:
        x.append(int(row[0]))
        selected.append(float(row[1]))

# Calculate ratio
ratio = [selected[i] / x[i] for i in range(len(x))]

# Plot
fig, ax1 = plt.subplots(figsize=(10,6))

# Left axis → selected activities
ax1.plot(x, selected, marker='o', linewidth=2, label="Selected Activities")
ax1.set_xlabel("Total Activities (n)")
ax1.set_ylabel("Number of Activities Selected")

# Right axis → ratio
ax2 = ax1.twinx()
ax2.plot(x, ratio, linestyle='--', linewidth=2, label="Selection Ratio")
ax2.set_ylabel("Selection Ratio")

# Title & grid
plt.title("Activity Selection Efficiency Analysis")
ax1.grid(True, linestyle='--', alpha=0.6)

# Combine legends
lines_1, labels_1 = ax1.get_legend_handles_labels()
lines_2, labels_2 = ax2.get_legend_handles_labels()
ax1.legend(lines_1 + lines_2, labels_1 + labels_2)

plt.tight_layout()
plt.show()
```

```python
import csv
import matplotlib.pyplot as plt

x = []
sparse = []
dense = []

with open("sparse_dense_ratio.csv", "r") as file:
    reader = csv.reader(file)
    next(reader)

    for row in reader:
        x.append(int(row[0]))
        sparse.append(float(row[1]))
        dense.append(float(row[2]))

plt.figure(figsize=(10,6))

plt.plot(x, sparse, marker='o', linewidth=2, label="Sparse Activities")
plt.plot(x, dense, marker='s', linewidth=2, linestyle='--', label="Dense Activities")

plt.xlabel("Total Activities (n)")
plt.ylabel("Selection Ratio")
plt.title("Sparse vs Dense Activity Selection Efficiency")

plt.legend()
plt.grid(True, linestyle='--', alpha=0.6)

plt.tight_layout()
plt.show()
```
