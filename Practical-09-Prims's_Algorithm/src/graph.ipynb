```python
import csv
import matplotlib.pyplot as plt

x = []
y = []
n2 = []

with open("prims_n2_analysis.csv", "r") as file:
    reader = csv.reader(file)
    next(reader)

    for row in reader:
        n = int(row[0])
        t = float(row[1])

        x.append(n)
        y.append(t)
        n2.append(n * n)

# Normalize n^2 curve
scale = (y[-1] / n2[-1]) * 1.5
n2 = [val * scale for val in n2]

plt.style.use('seaborn-v0_8')
plt.figure(figsize=(10,6))

# 🔵 Measured time (deep blue)
plt.plot(x, y, marker='o', linewidth=2.5, color="#5A2EAB", label="Measured Time")

# 🟠 Upper bound (orange dashed)
plt.plot(x, n2, linestyle='--', linewidth=2.5, color="#E81010", label="Upper Bound (n²)")

plt.xlabel("Number of Vertices (n)")
plt.ylabel("Execution Time (seconds)")
plt.title("Prim's Algorithm (O(n²)) Performance Analysis")

plt.legend()
plt.grid(True, linestyle='--', alpha=0.6)

plt.tight_layout()
plt.show()
```

```python
import csv
import matplotlib.pyplot as plt

x = []
md = []
ms = []
hd = []
hs = []

with open("prims_full_comparison.csv", "r") as file:
    reader = csv.reader(file)
    next(reader)

    for row in reader:
        x.append(int(row[0]))
        md.append(float(row[1]))
        ms.append(float(row[2]))
        hd.append(float(row[3]))
        hs.append(float(row[4]))

plt.style.use('seaborn-v0_8')
plt.figure(figsize=(10,6))

plt.plot(x, md, marker='o', linewidth=2.5, label="Matrix Dense", color='#E63946')
plt.plot(x, ms, marker='s', linewidth=2.5, label="Matrix Sparse", color='#F4A261')

plt.plot(x, hd, marker='^', linewidth=2.5, label="Heap Dense", color='#457B9D')
plt.plot(x, hs, marker='d', linewidth=2.5, label="Heap Sparse", color='#2A9D8F')

plt.xlabel("Number of Vertices (n)")
plt.ylabel("Execution Time (seconds)")
plt.title("Prim's Algorithm: Dense vs Sparse (Matrix vs Heap)")

plt.legend()
plt.grid(True, linestyle='--', alpha=0.6)

plt.tight_layout()
plt.show()
```
