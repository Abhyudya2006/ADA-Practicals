```python
#config 1 , test csv file 
#define START_SIZE 10000 
#define END_SIZE 1000000 
#define STEP_SIZE 5000 
#define SEARCHES 3000 
#define MAX_VALUE (2*END_SIZE)
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Load CSV
df = pd.read_csv("linear_time_test.csv")

n = df["ArraySize"].values
t = df["AvgTime"].values

# Choose c slightly above the maximum observed slope
c = max(t / n) * 1.05   # small safety margin (> max slope)

upper_bound = c * n

# Find n0: first n after which T(n) <= c*n holds for all later points
n0 = None
for i in range(len(n)):
    if np.all(t[i:] <= upper_bound[i:]):
        n0 = n[i]
        break

# Plot
plt.figure(figsize=(10, 6))
plt.plot(n, t, label="Measured T(n)")
plt.plot(n, upper_bound, '--', label="c  n (Upper Bound)")

plt.xlabel("Array Size (n)")
plt.ylabel("Time (seconds)")
plt.title("Experimental Proof of O(n) for Linear Search")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

# Print results
print(f"Chosen constant c  = {c:.6e}")
print(f"Estimated n0       = {n0}")
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load CSV files
hits_df = pd.read_csv("linear_hits_test.csv")
miss_df = pd.read_csv("linear_misses_test.csv")

# Use the last row (largest array size)
hits = hits_df.iloc[-1]["Hits"]
misses = miss_df.iloc[-1]["Misses"]

labels = ["Hits", "Misses"]
sizes = [hits, misses]

plt.figure()
plt.pie(
    sizes,
    labels=labels,
    autopct="%1.1f%%",
    startangle=90
)

plt.title("Linear Search: Hit vs Miss Distribution (Average Case)")
plt.axis("equal")  # Makes pie circular
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load CSV generated by C program
df = pd.read_csv("linear_avg_index3.csv")

# Extract data
n = df["ArraySize"]
avg_index = df["AvgFoundIndex"]

plt.figure(figsize=(10, 6))

# Measured average index (thicker line, subtle markers)
plt.plot(
    n, avg_index,
    marker='o',
    markersize=4,
    linewidth=2,
    label="Measured Avg Found Index"
)

# Theoretical n/2 line (clean dashed line)
plt.plot(
    n, n / 2,
    linestyle='--',
    linewidth=2,
    label="Theoretical n / 2"
)

plt.xlabel("Array Size (n)", fontsize=11)
plt.ylabel("Average Found Index", fontsize=11)
plt.title("Average Case Validation of Linear Search\n(Avg Found Index  n / 2)", fontsize=13)

plt.legend()
plt.grid(True, linestyle=':', linewidth=0.8)
plt.tight_layout()
plt.show()
```
