```python
import csv
import matplotlib.pyplot as plt

x = []
y = []
n2 = []

with open("dijkstra_performance.csv", "r") as file:
    reader = csv.reader(file)
    next(reader)

    for row in reader:
        n = int(row[0])
        t = float(row[1])

        x.append(n)
        y.append(t)
        n2.append(n * n)

# Normalize n^2 curve
scale = (y[-1] / n2[-1]) * 1.5
n2 = [val * scale for val in n2]

plt.style.use('seaborn-v0_8')
plt.figure(figsize=(10,6))

# 🔵 Measured Time
plt.plot(x, y, marker='o', linewidth=2.5, color='#2E86AB', label="Measured Time")

# 🔴 Upper Bound
plt.plot(x, n2, linestyle='--', linewidth=2.5, color='#E63946', label="Upper Bound (n²)")

plt.xlabel("Number of Vertices (n)")
plt.ylabel("Execution Time (seconds)")
plt.title("Dijkstra's Algorithm (O(n²)) Performance Analysis")

plt.legend()
plt.grid(True, linestyle='--', alpha=0.6)

plt.tight_layout()
plt.show()
```
