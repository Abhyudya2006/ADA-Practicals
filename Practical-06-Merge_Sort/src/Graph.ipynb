```python

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

def plot_mergesort_results():
    try:
        df = pd.read_csv("mergesort_avg_case.csv")
    except FileNotFoundError:
        print("Error: 'mergesort_avg_case.csv' not found.")
        return

    sizes = df["Size"].values
    times = df["Avg_Time_ms"].values

    # Offset for theoretical bound
    OFFSET = 1.2

    # --- Calculation of Theoretical Upper Bound (O(n log n)) ---
    c = (times[-1] / (sizes[-1] * np.log2(sizes[-1]))) * OFFSET
    bound_nlogn = c * (sizes * np.log2(sizes))

    plt.style.use("bmh")

    # --- Graph: Merge Sort ---
    plt.figure(figsize=(10, 6))
    
    plt.plot(sizes, times,
             color="#2ebbe2",
             marker='o',
             markersize=5,
             linewidth=2,
             label="Measured (Merge Sort)",
             zorder=3)

    plt.plot(sizes, bound_nlogn,
             color="#b10d0d",
             linestyle="--",
             linewidth=1.5,
             label=f"Upper Bound Ceiling ({OFFSET} * n log n)",
             zorder=2)

    # Complexity label near curve
    plt.text(sizes[-1], times[-1],
             r' $O(n \log n)$',
             color="#1f77b4",
             fontsize=14,
             fontweight='bold',
             va='center')

    plt.title("Merge Sort Performance: Average Case", fontsize=14, fontweight='bold')
    plt.xlabel("Array Size (n)")
    plt.ylabel("Time (milliseconds)")
    plt.legend()
    plt.grid(True, linestyle=":", alpha=0.6)
    plt.tight_layout()

    plt.show()


if __name__ == "__main__":
    plot_mergesort_results()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

def plot_mergesort_results():
    try:
        df = pd.read_csv("mergesort_avg_case_1.csv")
    except FileNotFoundError:
        print("Error: file not found.")
        return

    sizes = df["Size"].values
    times = df["Avg_Time_ms"].values

    OFFSET = 1.15
    c = (times[-1] / (sizes[-1] * np.log2(sizes[-1]))) * OFFSET
    bound_nlogn = c * (sizes * np.log2(sizes))

    # ✅ STYLE APPLIED ONLY INSIDE THIS BLOCK
    with plt.style.context("dark_background"):

        fig, ax = plt.subplots(figsize=(12, 7))

        neon_green = "#39FF14"
        dark_green = "#0F5107"

        ax.plot(sizes, times,
                color=neon_green,
                marker='o',
                markersize=4,
                linewidth=1.5,
                label="[SYSTEM] Measured (Merge Sort)",
                zorder=3)

        ax.plot(sizes, bound_nlogn,
                color=dark_green,
                linestyle="--",
                linewidth=1,
                label=f"[BOUND] Upper Ceiling ({OFFSET} * n log n)",
                zorder=2)

        ax.text(sizes[-1], times[-1],
                r' >> O(n log n)',
                color=neon_green,
                family='monospace',
                fontsize=12,
                fontweight='bold',
                va='center')

        ax.set_title("TERMINAL_PROCESS: MERGE_SORT_ANALYSIS",
                     color=neon_green,
                     fontsize=14,
                     fontweight='bold',
                     loc='left')

        ax.set_xlabel("> ARRAY_SIZE (n)", color=neon_green, family='monospace')
        ax.set_ylabel("> TIME (ms)", color=neon_green, family='monospace')

        legend = ax.legend(facecolor='black', edgecolor=neon_green)
        plt.setp(legend.get_texts(), color=neon_green)

        ax.grid(True, linestyle="--", alpha=0.2, color=neon_green)

        for spine in ax.spines.values():
            spine.set_edgecolor(dark_green)

        plt.tight_layout()
        plt.show()


if __name__ == "__main__":
    plot_mergesort_results()
```
