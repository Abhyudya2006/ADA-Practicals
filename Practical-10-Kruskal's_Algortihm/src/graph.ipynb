```python
import csv
import matplotlib.pyplot as plt
import math

x = []
y = []
elog = []

with open("kruskal_performance.csv", "r") as file:
    reader = csv.reader(file)
    next(reader)

    for row in reader:
        n = int(row[0])
        e = int(row[1])
        t = float(row[2])

        x.append(n)
        y.append(t)

        # E log E complexity
        elog.append(e * math.log2(e))

# Normalize curve
scale = (y[-1] / elog[-1]) * 1.5
elog = [val * scale for val in elog]

plt.style.use('seaborn-v0_8')
plt.figure(figsize=(10,6))

# 🔵 Measured time
plt.plot(x, y, marker='o', linewidth=2.5, color='#2E86AB', label="Measured Time")

# 🟠 Upper bound
plt.plot(x, elog, linestyle='--', linewidth=2.5, color='#F18F01', label="Upper Bound (E log E)")

plt.xlabel("Number of Vertices (n)")
plt.ylabel("Execution Time (seconds)")
plt.title("Kruskal's Algorithm Performance Analysis")

plt.legend()
plt.grid(True, linestyle='--', alpha=0.6)

plt.tight_layout()
plt.show()
```

```python
import csv
import matplotlib.pyplot as plt

x = []
dense = []
sparse = []

with open("kruskal_sparse_dense.csv", "r") as file:
    reader = csv.reader(file)
    next(reader)

    for row in reader:
        x.append(int(row[0]))
        dense.append(float(row[1]))
        sparse.append(float(row[2]))

plt.style.use('seaborn-v0_8')
plt.figure(figsize=(10,6))

# 🔴 Dense
plt.plot(x, dense, marker='o', linewidth=2.5, color='#E63946', label="Dense Graph")

# 🟢 Sparse
plt.plot(x, sparse, marker='s', linewidth=2.5, color='#2A9D8F', label="Sparse Graph")

plt.xlabel("Number of Vertices (n)")
plt.ylabel("Execution Time (seconds)")
plt.title("Kruskal's Algorithm: Dense vs Sparse Graph Performance")

plt.legend()
plt.grid(True, linestyle='--', alpha=0.6)

plt.tight_layout()
plt.show()
```
